## **Read Me**

Oryza sativa L. and Oryza glaberrima Steud.

### **Traits observed**
**Plant Type** 

Culm Angle - rice direct-seeded in rows.

**Leaf blade length**

Measure the highest leaf below the flag leaf (penultimate leaf) on the main culm from the ligule to the tip of the blade on five representative plants. Calculate average to nearest centimeter.

**Leaf blade width**

Measure at the widest portion of the penultimate leaf (i.e. highest leaf below the flag leaf) on the main culm on five representative plants. Calculate average to nearest cm. Oryza sativa leaf can be coded as narrow (< 1 cm), internediate (1-2 cm) or broad (>2 cm).

**Panicle length**

Length of main axis of panicle measured from the panicle base to the tip. Record the average of five representative plants. Stage: 7 days after anthesis or upon full panicle exsertion.

**Ligule length**

Ligule length in millimeter

**Ligule Shape**

**Ligule shape**

**Culm Diameter**

Culm diameter of basal internode at reproductive in millimeter

**Culm length**

Measured from ground level to the base of the panicle. Record the average of five actual measurements, to the nearest cm.

**Total number of tillers**

Count tillers with panicles and those without panicles on 5 samples

**Number of fertile tillers**

Count only the tillers with panicles on 5 samples.

## **The analisys:**

import the libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

uploading data:

```python
data=pd.read_csv('Data/Data.csv')
```

confering traits of columns:

```python
data
```

checking our data it is possible to see that some columns have a comma to separate the decimals and this will make them be read as a string

```python
data.describe()
```

then we will select the numeric columns of the dataframe and then apply the following code:

```python
cols= ['Leaf blade length [cm]', 'Leaf blade width [cm]',
       'Culm length [cm]', 'Panicle length [cm]', 'Ligule length [mm]',
       'Culm diameter [mm]','Total number of tillers ', 'Number of fertile tillers ']

data[cols]=data[cols].replace(',','.', regex=True)
data[cols]=data[cols].astype(float)
```

when describing our data again you can see that it selects all the numeric columns now

```python
data.describe()
```

At this point we need to transform the columns of our interest into a dataframe to create our graphs without major problems.

```python
df=data[cols]
df = df.apply(pd.to_numeric, errors='coerce')
df
```

It is always very interesting at the beginning of a multivariate analysis to apply point data to check the distribution of the data, see outliers and predictability.

```python
sns.pairplot(data[cols], diag_kind='kde')
plt.figure(figsize=(12,6))
plt.show()
```

A quick look at this matrix of graphs shows how the relationship between "Total number of tillers" and "Number of fertile tillers" distributes its values ââaround an imaginary curve, which is a clear sign that the data is correlated.

If in doubt, you can apply Pearson's correlation

```python
corr, pval = pearsonr(df['Total number of tillers '], df['Number of fertile tillers '])
print(f"Pearson correlation: {corr:.2f} (p-value: {pval:.6f})")
```

We can see that they have a positive correlation (if one increases the other increases too) of 77% and a p-value of 0 (meaning that there is no probability of this being due to chance)

```python
df = data[['Total number of tillers ', 'Number of fertile tillers ']].dropna()

sns.jointplot(
    data=df,
    x='Total number of tillers ', 
    y='Number of fertile tillers ',
    kind='reg',
    scatter_kws={'alpha':0.3, 'color':'#6666ff'},
    line_kws={'color':'#FF4D00', 'linewidth':1},
    
    
)

plt.xlabel('Total number of tillers ')
plt.xticks(np.arange(0,90,10))
plt.ylabel('Number of fertile tillers ')
plt.yticks(np.arange(0,110,10))
plt.grid(True, zorder=1)
plt.suptitle('Relationship between Vegetative Capacity and Productive Potential', fontsize=14)
plt.show()
```

When viewing the information in this graph, it becomes intuitive to see that rice reaches its maximum production capacity between 10-20 seeds, as this is where maximum vegetative use is concentrated.

based on this, we can ask ourselves 'what other characteristics are correlated' and we can create a 'heatmap' that answers us graphically

```python
df=data[cols]
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(axis=0, inplace=True)
df
```

```python
pearson_corr=df.corr(method='pearson')

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0)
plt.title("Pearson's correlation for the characteristics of Oryza sativa")
plt.show()
```

Among the other characteristics that have a strong positive correlation, we can highlight 'culm length (cm)' with 'Leaf blade length (cm), 'leaf blade width (cm)' with 'leaf blade length (cm)' and 'culm length (cm) with leaf blade width (cm)'.

This leads us to the next notes, as if by the Simpsons paradox, if together these analyzed characteristics can present a distinct correlation.

Now we can perform the moment we use machine learning using the sklearn library where we have several resources to model results based on our data.

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

First, we need to list the columns (X) that we will use to predict the result we want (y).

We must also remember that only numerical results are listed here.

```python
X=df[['Leaf blade length [cm]', 'Leaf blade width [cm]', 'Culm length [cm]',
       'Panicle length [cm]', 'Ligule length [mm]', 'Culm diameter [mm]',
       'Total number of tillers ']]
y=df['Number of fertile tillers ']
```

We separate 30% of the data to test and verify if our predictions work with the size of our dataframe and if this information is enough to predict 'y'

```python


# Suponha que X sejam os dados de entrada e y os rÃ³tulos/valores alvo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

```

We will be using a simple linear regression

```python
lm=linear_model.LinearRegression()
```

we can check the size of the data we used to train the model with the '.shape' command and we see that 315 rows were used

```python
X_train.shape, y_train.shape
```

we train the model and see the predictions that 'X_test' predicted

```python
lm.fit(X_train,y_train)
pred = lm.predict(X_test)
pred.shape
```

the 'lm.score' command gives us the variation of the data, with values ââcloser to 1 it means that the model explains well, with a value equal to 0.5 it is as predictable as if it were guessed by chance, a value of 0 means that the model does not explain any of the data

```python
lm.score(X_train,y_train)
```

However, 'lm.score' is not ideal because it tells us percentages and not values ââof our variable. For this we can use RMSE, the root mean square error, which shows us how much our model varies from the observations in terms of numbers of our variable.

Here 3.12 seeds

```python
mse = mean_squared_error(y_test, pred)
rmse=np.sqrt(mse)
print("RMSE:", rmse)
```

We can plot this on a graph and see more clearly that as we increase the number of fertile profiles we can see that there is a greater error

and where the data is more reliable, between 10 and 20 profiles

```python
# Certifique-se de que y_teste e y_pred sejam arrays numpy
y_teste = np.array(y_test)
y_pred = np.array(pred)

plt.figure(figsize=(6, 6))

# Plotagem principal
sns.regplot(
    x=y_test, 
    y=pred, 
    scatter_kws={'alpha': 0.3, 'color': '#6666ff'},
    line_kws={'color': '#FF4D00', 'linewidth': 1}
)

# Linha de 45 graus
plt.plot([y_teste.min(), y_teste.max()], [y_teste.min(), y_teste.max()], 'g--', label='Linha 45Â° (y = x)')

# Adicionando as distÃ¢ncias perpendiculares (resÃ­duos visuais)
for xi, yi in zip(y_teste, y_pred):
    # Projetar ponto (xi, yi) na reta y = x
    proj = (xi + yi) / 2
    plt.plot([xi, proj], [yi, proj], 'gray', alpha=0.5)

plt.xlabel("NÃºmero observado de perfilhos fÃ©rteis (y_teste)")
plt.ylabel("NÃºmero previsto de perfilhos fÃ©rteis (Å·)")
plt.title("ValidaÃ§Ã£o da RegressÃ£o Linear em Oryza spp.")
plt.grid(True)
plt.axis("equal")
plt.legend()
plt.show()
```

we can summarize everything that was seen by importing the 'statsmodels.api' library

```python
import statsmodels.api as sm
x_sm = sm.add_constant(X)
sm_model = sm.OLS(y, x_sm).fit()
```

```python
sm_model.summary()
```

Linear regression explained 83.5% of the variance in the variable "Number of fertile tillers" (RÂ² = 0.835).

-A 1 mm increase in Ligule length is associated with a 1.4419 unit increase in Number of fertile tillers

-Each additional unit in Total number of tillers is associated with a 0.7352 unit increase in Number of fertile tillers
